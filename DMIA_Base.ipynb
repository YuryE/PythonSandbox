```
import random

def function_1_1(my_list, n):
    summ=0
    for i in my_list:
        if i<=n:
            summ+=i
    return summ
    
random.seed(42)
print(function_1_1([random.randint(0, 1000) for _ in range(10000)], 700))

#function_1_1([1, 2, 3, 4, 5, 6, 3, 7], 7) == 31 OK

#function_1_1([1, 2, 3, 4, 4, 3, 2, 1], 7) == 20 OK

#function_1_1([1, 2, 3, 4, 4, 3, 2, 1], 3) == 12 OK

#function_1_1([1, 2, 3, 4, 5, 6, 3, 7], 5) == 18 OK

#function_1_1([1, 2, 3, 4, 5, 6, 3, 7], 3) == 9 OK

#function_1_1([1, 2, 3, 4, 5, 6, 3, 7], 0) == 0 OK
```

```
import collections
import random
import numpy as np

def function_1_2(my_list, dept):
  depts = collections.defaultdict(int)
  for surname, sigle_dept in my_list:
    depts[surname] += sigle_dept
    depts_sum = depts.items()
    final_list=sorted([(surname) for (surname,sigle_dept) in depts_sum if sigle_dept > dept])
  return final_list


random.seed(42)
print(','.join(function_1_2([
    (str(random.randint(1, 20)), random.randint(1, 1000) )
    for _ in range(1000)
], 25000)))

#function_1_2([('Ivanov', 1), ('Petrov', 2), ('Ivanov', 2), ('Sidorov', 4)], 1) == ['Ivanov', 'Petrov', 'Sidorov'] OK

#function_1_2([('Ivanov', 1), ('Petrov', 2), ('Ivanov', 2), ('Sidorov', 4)], 2) == ['Ivanov', 'Sidorov'] OK

#function_1_2([('Ivanov', 1), ('Petrov', 2), ('Ivanov', 2), ('Sidorov', 4)], 3) == ['Sidorov'] OK

#function_1_2([('Ivanov', 1), ('Petrov', 2), ('Ivanov', 2), ('Sidorov', 4)], 4) == [] OK

#function_1_2([('Ivanov', 1), ('Petrov', 2), ('Ivanov', 1), ('Sidorov', 4)], 2) == ['Sidorov'] OK

#function_1_2([('Ivanov', 1), ('Petrov', 2), ('Ivanov', 2), ('Petrov', 4)], 2) == ['Ivanov', 'Petrov'] OK

#function_1_2([('Ivanov', 1), ('Petrov', 2), ('Ivanov', 2), ('Petrov', 4)], 3) == ['Petrov'] OK
```

```
def function_2_1(n):
  factor=1
  for i in range(n):
    factor*=i+1
  return factor 

print(function_2_1(100))
#function_2_1(3) == 6 OK

#function_2_1(5) == 120 OK

#function_2_1(9) == 362880 OK

#function_2_1(13) == 6227020800 OK

```

```
import numpy as np
import random

def function_2_2(list1,list2):
  result=[]
  if np.mean(list1) < np.mean(list2):
    result.append('First')
  elif np.mean(list1)== np.mean(list2):
    result.append('Equal')
  elif np.mean(list1)> np.mean(list2):
    result.append('Second')
  if min(list1) < min(list2):
    result.append('First')
  elif min(list1)== min(list2):
   result.append('Equal')
  elif min(list1)> min(list2):
   result.append('Second')
  if max(list1) < max(list2):
    result.append('First')
  elif max(list1)== max(list2):
    result.append('Equal')
  elif max(list1)> max(list2):
    result.append('Second')
  print(np.mean(list1))
  print(min(list1))
  print(max(list1))
  print(np.mean(list2))
  print(min(list2))
  print(max(list2))
  return result
  
random.seed(47)
print(','.join(function_2_2(
    [random.randint(0, 100000) for _ in range(1000)],
    [random.randint(0, 100000) for _ in range(1000)]
)))

#function_2_2([1, 2, 3], [2, 2, 2]) == ('Equal', 'First', 'Second') OK

#function_2_2([2, 2, 1], [4]) ==('First', 'First', 'First') OK

#function_2_2([3, 5], [2, 4, 6]) == ('Equal', 'Second', 'First') OK

#function_2_2([1, 7], [2, 4, 7]) == ('First', 'First', 'Equal') OK

```

```
import random

def function_3_1(my_list,n,k):
  sum=0
  final_list=sorted([item for item in my_list if item <= n], reverse=True)
  for i in range(k):
    if i < len(final_list):
      sum+=final_list[i]
  return sum
  
random.seed(42)
print(function_3_1([random.randint(0, 1000) for _ in range(10000)], 700, 300))

#function_3_1([1, 2, 3, 4, 5, 6, 3, 7], 7, 1) == 7

#function_3_1([1, 2, 3, 4, 5, 6, 3, 7], 7, 3) == 18

#function_3_1([1, 2, 3, 4, 5, 6, 3, 7], 7, 5) == 25

#function_3_1([1, 2, 3, 4, 5, 6, 3, 7], 7, 8) == 31

#function_3_1([1, 2, 3, 4, 5, 6, 3, 7], 7, 10) == 31

#function_3_1([1, 5, 3, 6, 5, 6, 3, 6], 6, 3) == 18

#function_3_1([1, 5, 3, 6, 5, 6, 3, 6], 5, 3) == 13

#function_3_1([1, 5, 3, 6, 5, 6, 3, 6], 4, 3) == 7

#function_3_1([1, 5, 3, 6, 5, 6, 3, 6], 6, 2) == 12

#function_3_1([1, 5, 3, 6, 5, 6, 3, 6], 5, 2) == 10

#function_3_1([1, 5, 3, 6, 5, 6, 3, 6], 4, 2) == 6

```

```
def function_3_2(n):
    primes=[]
    for p in range(1, n+1):
      for x in range(1, p+1):
        if p % x != 0 or x==1:
          continue
        elif p==x:
          primes.append(p)
          break
        elif p % x == 0:
          break
    return primes

print(sum(function_3_2(10000)))
#function_3_2(3) == [2, 3]

#function_3_2(10) == [2, 3, 5, 7]

#function_3_2(20) == [2, 3, 5, 7, 11, 13, 17, 19]



```

```
import pandas as pd

bioresponce = pd.read_csv(r"https://raw.githubusercontent.com/data-mining-in-action/DMIA_Base_2019_Autumn/master/selection/bioresponse.csv", header=0, sep=',')

bioresponce.head(5)

y = bioresponce.Activity.values
X = bioresponce.iloc[:, 1:]



from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier

from sklearn.model_selection import cross_val_score

models = [
    LogisticRegression(),
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    LinearSVC(),
    RandomForestClassifier(n_estimators=100), 
    GradientBoostingClassifier(n_estimators=100)
]
```

```
for model in models:
    %time print(model, '\nAccuracy:', cross_val_score(model, X, y, cv=5).mean())
    print()



from sklearn.model_selection import GridSearchCV


    #Параметр C по степеням 10 от -2 до +2
    #Параметр max_iter от 100 до 500 включительно с шагом 100
    #Параметр penalty либо 'l1' либо 'l2'
    #Параметр random_state обязательно 42
    #Параметр fit_intercept либо False либо True

param_grid={'C':[10**i for i in range(-2,3)],'max_iter':[i for i in range(100,501,100)], 'penalty':['l1', 'l2'],'random_state':[42],'fit_intercept':[True,False]}

print(round(GridSearchCV(
   LogisticRegression(),  param_grid, cv=5
).fit(X, y).best_score_, 4))
```
