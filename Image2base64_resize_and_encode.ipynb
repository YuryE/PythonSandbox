```python
from pathlib import Path as ps
# определение пути
def get_files(directory, files=[]):
    currentDirectory = ps(directory)
    for currentFile in currentDirectory.iterdir():
        if currentFile.is_file():
            files.append(currentFile)
    return files # Формирует список Windowpath-объектов

files=get_files('C:\projects\sm-code\shop\EXE2\Images') # Принимает полный адрес директории
```

```python
from PIL import Image
def scale_image(input_image_path,
                output_image_path,
                width=None,
                height=None
                ):
    original_image = Image.open(input_image_path)
    w, h = original_image.size
    print('The original file: '+input_image_path)
    print('The original image size is {wide} wide x {height} '
          'high'.format(wide=w, height=h))
 
    if width and height:
        max_size = (width, height)
    elif width:
        max_size = (width, h)
    elif height:
        max_size = (w, height)
    else:
        # No width or height specified
        raise RuntimeError('Width or height required!')
 
    original_image.thumbnail(max_size, Image.ANTIALIAS)
    original_image.save(output_image_path)
 
    scaled_image = Image.open(output_image_path)
    width, height = scaled_image.size
    print('The scaled file: '+output_image_path)
    print('The scaled image size is {wide} wide x {height} '
          'high'.format(wide=width, height=height))

def make_scaled_images (original_files,new_width=None,new_heght=None,scaled_files=[]):    
    for i in original_files:
        scaled_file=ps(i.parent,i.stem+'2'+i.suffix)
        if __name__ == '__main__':
            scale_image(input_image_path=str(i),
                    output_image_path=str(scaled_file),
                    width=new_width,
                    height=new_heght)
        scaled_files.append(scaled_file)
    return scaled_files

scaled_files=make_scaled_images(files,80) # Принимает список Windowpath-объектов
```

```python
import base64 as bs    
 
def base64_encode_images(images):
    for i in images:
        image_read = i.read_bytes() #open binary file in read mode
        image_64_encode = bs.b64encode(image_read)
        base64=str((image_64_encode))[2:-1]
        filename=(i.stem[:-1])+i.suffix #Отрезаю у нового файла постфикс
        print("""declare
    l_clob      clob;
    l_filename  varchar2(254);
begin  
    l_clob := '"""+base64+"""';
    l_filename := '"""+filename+"""';

    MERGE INTO t_shp_ware_epl_image p
      USING ( SELECT l_clob     as image64
                   , l_filename as filename
                FROM dual
            ) p1 ON ( p.filename = p1.filename )
      WHEN MATCHED THEN UPDATE SET p.image64 = p1.image64
      WHEN NOT MATCHED THEN INSERT ( p.filename , p.image64)
                            VALUES ( p1.filename, p1.image64);
    commit;                            
end;
/
""")
        
base64_encode_images(scaled_files)
```

```python
def remove_files(files):
    for i in files:
        i.unlink
remove_files(scaled_files)
print(scaled_files)
```
