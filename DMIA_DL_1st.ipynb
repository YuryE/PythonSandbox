```
import keras
import matplotlib.pyplot as plt
import numpy as np

from numpy.random import seed
from tensorflow import set_random_seed


import pkg_resources
pkg_resources.get_distribution("keras").version

seed(1)
set_random_seed(2)

(x_train, y_train), (x_test, y_test) = keras.datasets.cifar10.load_data()
print('Answer #1:'+str(x_train.shape[0] + y_train.shape[0]))

class_names = {0: 'airplane', 1: 'automobile', 2: 'bird', 3: 'cat', 4: 'deer',
               5: 'dog', 6: 'frog', 7: 'horse', 8: 'ship', 9: 'truck'}

fig = plt.figure(figsize=(8,3))

for i in range(len(class_names)):
    ax = fig.add_subplot(2, 5, 1 + i, xticks=[], yticks=[])
    ax.set_title(class_names[i])
    idx = np.where(y_train[:]==i)[0][0]
    image = x_train[idx]
    plt.imshow(image)

plt.show()

from IPython.display import SVG
from sklearn.metrics import accuracy_score
from keras.utils.vis_utils import model_to_dot
from keras.models import Sequential
from keras.layers import (Dense, Conv2D, Activation,
                          Dropout, MaxPooling2D, Flatten)
from keras.optimizers import Adam
from keras.utils import to_categorical



def create_model():
    ##########################################
    created_model = Sequential
    ##########################################
    return created_model

def add_blocks_to_model(created_model):
    
    created_model.add(Conv2D(filters=32,
                         kernel_size=(3, 3),
                         padding='same',
                         input_shape=x_train.shape[1:]),1)
    created_model.add(Activation('relu'))
    
    ##########################################
    #created_model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same', input_shape=x_train.shape[1:]))
    created_model.add(Activation(activation='relu'))
    created_model.add(Conv2D(filters=32, kernel_size=(3, 3), padding='same'))
    created_model.add(Activation(activation='relu'))
    created_model.add(MaxPooling2D(pool_size=(2, 2)))
    created_model.add(Dropout(rate=0.25))
    created_model.add(Conv2D(filters=64, kernel_size=(3, 3), padding='same'))
    created_model.add(Activation(activation='relu'))
    created_model.add(Conv2D(filters=64, kernel_size=(3, 3), padding='same'))
    created_model.add(Activation(activation='relu'))
    created_model.add(MaxPooling2D(pool_size=(2, 2)))
    created_model.add(Dropout(rate=0.25))
    created_model.add(Flatten())
    created_model.add(Dense(units=128))
    created_model.add(Activation(activation='relu'))
    created_model.add(Dropout(rate=0.5))
    created_model.add(Dense(units=10))
    created_model.add(Activation(activation='softmax', name='out'))
    ##########################################
    
    created_model.add(Flatten())
    created_model.add(Dense(units=128))
    created_model.add(Activation('relu'))
    created_model.add(Dropout(0.5))
    created_model.add(Dense(units=10))
    created_model.add(Activation('softmax', name='out'))
    
    return created_model
                      
model = create_model()
model = add_blocks_to_model(model)
model.summary()
SVG(model_to_dot(model).create(prog='dot', format='svg'))

```
